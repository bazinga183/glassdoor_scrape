```python
import re
import pandas as pd
import numpy as np
```

```python
ms_reviews = pd.read_csv('Company Reviews\Microsoft Reviews.csv')

ms_reviews
```

```python
average_employee_rating = round(ms_reviews['Rating'].mean(), 2)
average_work_life_balance = round(ms_reviews['Work-Life Balance'].mean(), 2)
average_culture_value = round(ms_reviews['Culture and Values'].mean(), 2)
average_diversity_inclusion = round(ms_reviews['Diversity and Inclusion'].mean(), 2)
average_career_opportunities = round(ms_reviews['Career opportunities'].mean(), 2)
average_compensation_benefits = round(ms_reviews['Compensation and Benefits'].mean(), 2)
average_senior_management = round(ms_reviews['Senior Management'].mean(), 2)

average_ratings = [average_employee_rating,
                  average_work_life_balance,
                  average_culture_value,
                  average_diversity_inclusion,
                  average_career_opportunities,
                  average_compensation_benefits,
                  average_senior_management]

ratings_df = pd.DataFrame([average_ratings], columns = ['Rating', 'Work/Life Balance',
                                                     'Culture & Value', 'Diversity & Inclusion',
                                                     'Career Opportunities', 'Compensation & Benefits',
                                                     'Senior Management'])

ratings_df
```
